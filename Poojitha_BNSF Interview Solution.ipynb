#####  Regardless of prior information about data, it is observed that given data is collected between March 29th, 2019(week 13) and September 1st, 2019(week 35). The data represents number of events occured with respect to sw_version, splits and split_values in a given time period of time. There are 2 sw_versions, 4 splits, 33 split_values repersenting characteristics with Nominal data type. 

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

sns.set()
```

```python
df= pd.read_csv("Data Scientist interview.csv", header='infer')
df.dtypes
df.shape
```

### DATA WRANGLING

```python
# Remove duplicates
df.drop_duplicates(inplace=True, keep='first')
df.shape
```

```python
# Search for null values
df.isna().sum()
```

### Trend Analysis

```python
pd.pivot_table(df, index='week', columns='split', values="events_count",aggfunc=np.sum)
#plt.ylabel("split by weeks")
```

```python
pd.pivot_table(df, index='week', columns='split', values="events_count",aggfunc=np.sum).plot(kind='bar', figsize=(8,6))
plt.ylabel("split by weeks")
```

### Bar chart shows the events count by split in 22 weeks. 
1. event count in split category target has not followed any trend during this time, no variation in events occured.
2. Remaining 3 splits has has followed unusual rise and fall in 25th and 32nd weeks. Rest of the time there is gradual increase and decreases of events during this period. 

### spit and split_value vs count

```python
# pivot table by spit and split_value vs count
pd.pivot_table(df, index=['split','split_value'], columns='week', values="events_count")


```

```python
#plot split type 'impactful'

fig, ax = plt.subplots(figsize=(10,6))
df.loc[df['split']=='impactful'].groupby(['week','split_value'])['events_count'].agg(np.sum).unstack().plot(ax=ax)
ax.set_xlabel('week')
ax.set_ylabel('Number of events')
```

```python
#plot split type 'manageable'
fig, ax = plt.subplots(figsize=(10,6))
df.loc[df['split']=='manageable'].groupby(['week','split_value'])['events_count'].agg(np.sum).unstack().plot(ax=ax)
ax.set_xlabel('week')
ax.set_ylabel('Number of events')
```

```python
#plot split type 'item_module'
fig, ax = plt.subplots(figsize=(15,8))
df.loc[df['split']=='item_module'].groupby(['week','split_value'])['events_count'].agg(np.sum).unstack().plot(ax=ax)
ax.set_xlabel('week')
ax.set_ylabel('Number of events')
```

```python
#plot split type 'target'
fig, ax = plt.subplots(figsize=(15,8))
df.loc[df['split']=='target'].groupby(['week','split_value'])['events_count'].agg(np.sum).unstack().plot(ax=ax)
ax.set_xlabel('week')
ax.set_ylabel('Number of events')
```

### Line chart shows split,spli_values Vs  event_count
1. Split impactful contains only 2 split_values in the events during this period, where impactful bagged highest count in 25th week and non-impactful bagged highest count in 32nd week.
2. Split Manageable experirnced gradual increase and sudden fall on the count by the end the period.
3. Split item_module with 16 split_values matained less variability in the count of events occured throught the period.But spli_values 220R has experience a sudden rise and fall in 25th week.
4. Split target with 13 split_values, only 2(SW.UNK,SIG) had experienced gradual increase on events count. Rest of them are constant through the period.
